```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
df=pd.read_csv("/kaggle/input/car-crashes/car_crashes.csv")
df.columns
```

```python
#to drop abbrev column axis=1 to drop col axis =0
df=df.drop("abbrev",axis=1)
df
```

```python
df.describe()
```

```python
# Heat map describe correlation between each column to remain all colums
# it describe relation how to varibles are co-related each other 
# heare green are highly co-related and red is low co-related
#annot - - is for to show value in each cell
#cmap --- is for graph color
#fmt is for to control decimal values
sns.heatmap(df.corr(),annot=True,cmap="PiYG",fmt="0.2f")
```
